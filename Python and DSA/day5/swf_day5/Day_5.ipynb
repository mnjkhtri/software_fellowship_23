## Stack
Stack is a Last In First Out data structure.  
It is very similar to list with the added constraint that only the latest item added in the list can be accessed/removed.


We can actually use a list to implement a stack.  
We only have to take care that the insertions and deletions always happen at the end of list.  
The stack data structure specifies the *method of access* of elements along with the structure within the data itself.

````python
stack = [1, 2, 3]
stack.append(4)
stack.pop()
````

Let us use the OOP principles we learned in the last class to implement a simple Stack data structure

Lets define a *Stack* class. Each object of *Stack* class has an attribute named *stack*. This is where the data is actually stored.  

The *stack* attribute is nothing more than a list. We have used the constructor of class to give the list an initial value.  

Now let us add methods to push and pop elements from the stack. 


````python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)
        
    def pop(self):
        if len(self.stack)==0:
            print("Cannot pop, stack is empty!")
            return
        return self.stack.pop()

s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.pop()
s.pop()
````

### Stack use case
So where do we need stack.  
* How does your text editor keep track of all your edits and undos? One possible way is to push your changes to stack for each of your edit and pop these changes during undo. Notice how the last-in-first-out scheme fits nicely in this situation.
* The compiler uses stack to keep track of nested function calls. Each function call gets pushed to stack and returning from the function pops it from the stack. 
* And many more use cases. All of these use cases feature a need match some *opening action* with some *closing action*

# **Search Algorithms**

Any algorithm which solves the search problem, to retrieve information stored within some data structure is called a search algorithm.

# **Types of Search Algorithms**

# *Linear Search or Sequential Search*
This algorithm works by sequentially iterating through the whole array or list from one end until the target element is found. 




**Approach for Linear or Sequential Search**

1. Start with index 0 and compare each element with the target

2. If the target is found to be equal to the element, return its index

3. If the target is not found, return -1

````python
def linear_search(arr, target):
    for i in range(len(arr)):
        if target==arr[i]:
            return i
    #Unable to find the target
    return -1


#create a numbers list
arr = []
#create a target
target = 0 

print(linear_search(arr, target))
````

# *Binary Search*

This type of searching algorithm is used to find the position of a specific value contained in a sorted array.


**Approach for Binary Search**

1. Compare the target element with the middle element of the array.

2. If the target element is greater than the middle element, then the search continues in the right half.

3. Else if the target element is less than the middle value, the search continues in the left half.

4. This process is repeated until the middle element is equal to the target element, or the target element is not in the array.

5. If the target element is found, its index is returned, else -1 is returned.

````python
def binary_search(arr, target):
    low = 0
    high = len(arr)-1

    while low<=high:
        mid = (low+high)//2

        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            low = mid+1
        else:
            high = mid-1
        
    return -1

#create a number list
arr = [1,2,3,4,5,6]
#create a target
target = 3

print(binary_search(arr, target))
````

# *Bubble Sort*


````python
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr


numbers = [23, 41, 25, 54, 18, 14]

print(bubble_sort(numbers))
````
