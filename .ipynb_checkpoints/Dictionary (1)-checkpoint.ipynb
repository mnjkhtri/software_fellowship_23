# **Dictionary**

---

*   Generally known as an associative array
*   Associative Array: An abstract data type that stores a collection of ***(key, value) pairs***, such that each possible key appears at most once in the collection.

**How the dictionary is created?**

```python

}
```

**Underlying Concept of Dictionaries: Hashing**


1.   When you create a dictionary and add a key-value pair to it, Python calculates the hash value of the key using a hashing function.
2.   This hashing function takes the key and performs a series of mathematical operations to generate a unique numerical value. This value is typically an integer.
3.   The hash value is then used as an index to determine where in the computer's memory the corresponding value should be stored. This allows for efficient retrieval of values later on.
4. When you want to retrieve a value from the dictionary using a key, Python again calculates the hash value of the key and uses it to locate the stored value in memory.**This process is very fast** because it directly points to the memory location without searching through the entire dictionary.
5. If multiple keys happen to have the same hash value (known as a hash collision), Python uses additional techniques to handle it. One common technique is to create a linked list of values at that index, where each value is associated with a different key that resulted in the same hash value. This way, Python can still retrieve the correct value based on the matching key.

**Time Complexity in Dictionaries**


1.   Insertion (Adding a key-value pair): O(1)
2.   Deletion (Removing a key-value pair): O(1)
3.   Retrival (Accessing a value using a key): O(1)

**Iterating through dictionary items.**

---



````python
thisdict = {
     'Colorado' : 'Rockies',
     'Boston'   : 'Red Sox',
     'Minnesota': 'Twins',
     'Milwaukee': 'Brewers',
     'Seattle'  : 'Mariners'
}
````

Print all key names in the dictionary, one by one:

````python
for x in thisdict:
  print(x)
````

Print all values in the dictionary, one by one:

````python
for x in thisdict:
  print(thisdict[x])
````

You can also use the values() method to return values of a dictionary:

````python
for x in thisdict.values():
  print(x)
````

You can use the keys() method to return the keys of a dictionary:



````python
for x in thisdict.keys():
  print(x)
````

Loop through both keys and values, by using the items() method:

````python
for x, y in thisdict.items():
  print(x, y)
````


**Dictionary Comprehension**

---



---



`Syntax:`
```python
dictionary = {key: value for vars in iterable}
```
`If Conditional Dictionary Comprehension`
```python
dictionary = {key:value for vars in iterable if condition}
```

`Multiple if Conditional Dictionary Comprehension`

Here both conditions should be `True`.
```python
dictionary = {key:value for vars in iterable if condition1 if condition2}
```

**Examples**

---



````python
#item price in dollars
old_price = {'milk': 1.02, 'coffee': 2.5, 'bread': 2.5}

dollar_to_pound = 0.76
new_price = {item: value*dollar_to_pound for (item, value) in old_price.items()}
print(new_price)
````

````python
original_dict = {'jack': 38, 'michael': 48, 'guido': 57, 'john': 33}

even_dict = {k: v for (k, v) in original_dict.items() if v % 2 == 0}
print(even_dict)
````

````python
original_dict = {'jack': 38, 'michael': 48, 'guido': 57, 'john': 33}

new_dict = {k: v for (k, v) in original_dict.items() if v % 2 != 0 if v < 40}
print(new_dict)
````
